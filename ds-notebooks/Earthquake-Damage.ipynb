```python
damage = pd.read_csv('https://raw.githubusercontent.com/labs13-quake-viewer/ds-data/danielle-romanoff/earthquake_damage.csv')
damage.head()
```

```python
damage.shape
```

```python
damage.Magnitude.isnull().sum()
```

```python
damage = damage.drop(0)
damage.head()
```

```python
damage = damage.query('Year > 1677')
damage['Year'] = damage['Year'].astype(int).astype(str)
damage['Mo'] = damage['Mo'].fillna(1).astype(int).astype(str)
damage['Dy'] = damage['Dy'].fillna(1).astype(int).astype(str)
damage['Date'] = damage['Year'] + '/' + damage['Mo'] + '/' + damage['Dy']
damage['Date'] = pd.to_datetime(damage['Date'], infer_datetime_format=True)
damage.head()
```

```python
damage_used = damage.copy()
damage_used.head()
```

```python
damage_used = damage_used.drop(['Year', 'Mo', 'Dy', 'Volcano'], axis = 1)
damage_used.head()
```

```python
damage_used = damage_used.fillna(0.0)
damage_used.head()
```

```python
damage_used['Injuries/Scale'].value_counts()
```

```python
damage_used['Tsunami'] = np.where(damage_used.Tsunami == 'Tsu', 1, damage_used.Tsunami)
damage_used['Deaths/Scale'] = np.where(damage_used['Deaths/Scale'] == 1, 50, damage_used['Deaths/Scale'])
damage_used['Deaths/Scale'] = np.where(damage_used['Deaths/Scale'] == 2, 100, damage_used['Deaths/Scale'])
damage_used['Deaths/Scale'] = np.where(damage_used['Deaths/Scale'] == 3, 1000, damage_used['Deaths/Scale'])
damage_used['Deaths/Scale'] = np.where(damage_used['Deaths/Scale'] == 4, 5000, damage_used['Deaths/Scale'])
damage_used.head()
```

```python
damage_used['Missing/Scale'] = np.where(damage_used['Missing/Scale'] == 1, 50, damage_used['Missing/Scale'])
damage_used['Missing/Scale'] = np.where(damage_used['Missing/Scale'] == 2, 100, damage_used['Missing/Scale'])
damage_used['Missing/Scale'] = np.where(damage_used['Missing/Scale'] == 3, 1000, damage_used['Missing/Scale'])
damage_used['Missing/Scale'] = np.where(damage_used['Missing/Scale'] == 4, 5000, damage_used['Missing/Scale'])
```

```python
damage_used['Injuries/Scale'] = np.where(damage_used['Injuries/Scale'] == 1, 50, damage_used['Injuries/Scale'])
damage_used['Injuries/Scale'] = np.where(damage_used['Injuries/Scale'] == 2, 100, damage_used['Injuries/Scale'])
damage_used['Injuries/Scale'] = np.where(damage_used['Injuries/Scale'] == 3, 1000, damage_used['Injuries/Scale'])
damage_used['Injuries/Scale'] = np.where(damage_used['Injuries/Scale'] == 4, 5000, damage_used['Injuries/Scale'])
damage_used.head()
```

```python
damage_used['Damage/ Scale'] = np.where(damage_used['Damage/ Scale'] == 1, 500000, damage_used['Damage/ Scale'])
damage_used['Damage/ Scale'] = np.where(damage_used['Damage/ Scale'] == 2, 2500000, damage_used['Damage/ Scale'])
damage_used['Damage/ Scale'] = np.where(damage_used['Damage/ Scale'] == 3, 12500000, damage_used['Damage/ Scale'])
damage_used['Damage/ Scale'] = np.where(damage_used['Damage/ Scale'] == 4, 27500000, damage_used['Damage/ Scale'])
damage_used.head()
```

```python
damage_used['Houses Destroyed/ Scale'] = np.where(damage_used['Houses Destroyed/ Scale'] == 1, 50, damage_used['Houses Destroyed/ Scale'])
damage_used['Houses Destroyed/ Scale'] = np.where(damage_used['Houses Destroyed/ Scale'] == 2, 100, damage_used['Houses Destroyed/ Scale'])
damage_used['Houses Destroyed/ Scale'] = np.where(damage_used['Houses Destroyed/ Scale'] == 3, 1000, damage_used['Houses Destroyed/ Scale'])
damage_used['Houses Destroyed/ Scale'] = np.where(damage_used['Houses Destroyed/ Scale'] == 4, 5000, damage_used['Houses Destroyed/ Scale'])
damage_used.head()
```

```python
damage_used['Houses Damaged/ Scale'] = np.where(damage_used['Houses Damaged/ Scale'] == 1, 50, damage_used['Houses Damaged/ Scale'])
damage_used['Houses Damaged/ Scale'] = np.where(damage_used['Houses Damaged/ Scale'] == 2, 100, damage_used['Houses Damaged/ Scale'])
damage_used['Houses Damaged/ Scale'] = np.where(damage_used['Houses Damaged/ Scale'] == 2, 1000, damage_used['Houses Damaged/ Scale'])
damage_used['Houses Damaged/ Scale'] = np.where(damage_used['Houses Damaged/ Scale'] == 2, 5000, damage_used['Houses Damaged/ Scale'])
damage_used.head()
```

```python
damage_used['Deaths'] = damage_used['Deaths/Scale'] + damage_used['Deaths/Num']
damage_used['Missing'] = damage_used['Missing/Scale'] + damage_used['Missing/Num']
damage_used['Injuries'] = damage_used['Injuries/Scale'] + damage_used['Injuries/Num']
damage_used['Damage'] = damage_used['Damage/ Scale'] + damage_used['Damage/$Mill']
damage_used['Houses Destroyed'] = damage_used['Houses Destroyed/ Num'] + damage_used['Houses Destroyed/ Scale']
damage_used['Houses Damaged'] = damage_used['Houses Damaged/ Num'] + damage_used['Houses Damaged/ Scale']
damage_used.head()
```
